```python
iris_dataframe=pd.DataFrame(data=iris.data,columns=["SL","SW","PL","PW"])
iris_dataframe
```

```python
iris_dataframe["target"]=iris.target #adicionando target al dataframe
iris_dataframe
```

```python
y=iris_dataframe["SW"]
print(y.mean()) #valor promedio de las SW
col=iris_dataframe["target"]
```

```python
col=(iris_dataframe["target"]+1)*2
col
```

```python
PL=iris_dataframe["PL"]*50
```

```python
plt.scatter(x,y,c=col,s=PL,alpha=0.5) #s significa markersize para scatter y alpha para la transparencia y va de 0 a 1
plt.xlabel("Largo del sépalo (cm)")
plt.ylabel("Ancho del sépalo (cm)")
plt.title("Ancho de Sépalo vs Largo de Sépalo")
plt.grid(True)
plt.show()
```

```python
PW=iris_dataframe["PW"]*50
```

```python
plt.scatter(x,y,c=col,s=PW,alpha=0.75) 
plt.xlabel("Largo del sépalo (cm)")
plt.ylabel("Ancho del sépalo (cm)")
plt.title("Ancho de Sépalo vs Largo de Sépalo")
plt.grid(True)
plt.show()
```
