```python
import matplotlib.pyplot as plt
import numpy as np 
```

## CinÃ©tica de primer orden 
## A --> B

```python
#dar valor a la conc de A, B, dty k1
A=1.0
B=0
k1=0.1
dt=0.1
t=0
x=[]
yA=[]
yB=[]
#usar ciclo while para asegurar que el tiempo va aumentando
while t<100:
    x.append(t) #append para aÃ±adir objetos a la lista vacia creada
    yA.append(A)
    yB.append(B)
    dA=-k1*A*dt
    dB=+k1*A*dt
    A+=dA
    B+=dB
    t+=dt
print(yA)
```

```python
plt.plot(x,yA)
plt.plot(x,yB)
plt.xlabel("t(min)")
plt.ylabel("Concentracion")
plt.title("Cinetica de primer orden")
plt.show()
```

## CinÃ©tica secuencial
## A --> B
## B --> C

```python
A=1.0
B=0
C=0
k1=0.1
k2=0.1
dt=0.1
t=0
x=[]
yA=[]
yB=[]
yC=[]
while t<100:
    x.append(t) #append para aÃ±adir objetos a la lista vacia creada
    yA.append(A)
    yB.append(B)
    yC.append(C)
    dA=-k1*A*dt
    dB=-k2*B*dt+k1*A*dt
    dC=k2*B*dt
    A+=dA
    B+=dB
    C+=dC
    t+=dt
print(yC)
```

```python
plt.plot(x,yA)
plt.plot(x,yB)
plt.plot(x,yC)
plt.xlabel("t(min)")
plt.ylabel("Concentracion")
plt.title("Cinetica secuencial")
plt.show()
```

## Equilibrio de primer orden
## A < -- > B

```python
#lo mismo que en la primera, solo se cambian las formulas
A=1.0
B=0
k1=0.1
k2=0.2
dt=0.1
t=0
x=[]
yA=[]
yB=[]

while t<100:
    x.append(t) 
    yA.append(A)
    yB.append(B)
    dA=-k1*A*dt+k2*B*dt
    dB=+k1*A*dt-k2*B*dt
    A+=dA
    B+=dB
    t+=dt
```

```python
plt.plot(x,yA)
plt.plot(x,yB)
plt.xlabel("t(min)")
plt.ylabel("Concentracion")
plt.title("Equilibrio")
plt.show()
```

## Cinetica de segundo orden 
## A +A < -- > B

```python
A=1.0
B=0
k1=0.1
dt=0.1 #para calculos mas exactos se disminuye el dt
t=0
x=[]
yA=[]
yB=[]

while t<100:
    x.append(t) 
    yA.append(A)
    yB.append(B)
    dA=-k1*A*A*dt
    dB=+k1*A*A*dt
    A+=dA
    B+=dB
    t+=dt
plt.plot(x,yA)
plt.plot(x,yB)
plt.xlabel("t(min)")
plt.ylabel("Concentracion")
plt.title("Cinetica de segundo orden")
plt.show()
```
