```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
import requests
```

```python
#Coeficientes de SOmate para el Diazeno N2H2
A=8.8997
B=93.7967
C=-53.12099
D=12.1159
E=0.36272
H298=55.13 #kJ/mol
```

```python
#generar arreglo de temperaturas que va desde 298 K a 1400 K de 0.1 en 0.1
dT=0.1
T=np.arange(298,1400,dT)

t=T/1000

Cp=A+B*t+C*t**2+D*t**3+E/t**2   #su unidad es J/K-mol
Cp
#calcular delta H
#cuanto calor necesito para ir desde 0 a 1400?
H=np.sum(Cp*dT)

H=np.cumsum(Cp*dT)+55.13*1000 #as las temepraturas desde 298 a  1400
H
```

```python
#como calcular s en forma acumulativa
S=np.cumsum(Cp/T*dT)
G=H-T*S
```

```python
plt.plot(T,Cp,color="blue")
plt.title("Cp vs T")
plt.xlabel("T(K)")
plt.ylabel("Cp (J/K-mol)")
plt.show()
```

```python
#graficar H , S y G
plt.plot(T,H,color="blue")
plt.plot(T,S,color="red")
plt.plot(T,G,color="green")
plt.show()
```
