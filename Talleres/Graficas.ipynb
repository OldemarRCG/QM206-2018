```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.plot([0,2,4,6],[3,1,7,4],color="black",linewidth=3,alpha=0.9,marker="o",markersize=8,markeredgewidth=2,markeredgecolor="black",markerfacecolor="lightseagreen")
plt.xlabel("$x(cm^2)$")
plt.ylabel("$Intensidad(\mu)$")
plt.title("Experimento No. 1")
plt.grid(True)
plt.show()
```

```python
x=np.arange(0,2*np.pi,0.1)
x
y=np.sin(x)
y
plt.plot(x,y)
plt.show()
```

```python
x=np.arange(0,2*np.pi,0.1)
x
y=np.sin(2*x)
y
plt.plot(x,y)
plt.show()
```

```python
x=np.arange(0,2*np.pi,0.1)
x
y=np.sin(4*x)
y2=np.cos(2*x)
y
plt.plot(x,y,color="forestgreen")
plt.plot(x,y2,color="red",dashes=[2,2])
plt.title("Funcion sen(4x) 0-$2\pi$")
plt.xlabel("t(s)")
plt.ylabel("Amplitud de la onda")
plt.grid(True)
plt.show()
```

```python
x=np.random.rand(100)*10
y=np.random.rand(100)*10
r=np.random.rand(100)*250+50
plt.scatter(x,y,s=r,color="red")
plt.show()
```

```python
x=np.random.randn(10000)
x
plt.hist(x,bins=50,facecolor="lime",edgecolor="black",density=True)
plt.title("Distribución de la población")
plt.ylabel("Densidad habitantes/$m^2$")
plt.text(-3, .35,r'$\mu=100, \ \sigma=1.0$')
plt.show()
```
